# UAC Predictive Forecasting
Forecasting **Children in HHS Care** and **Discharge Demand** using statistical and machine-learning models,
following the same structured workflow used in the reference cancer-diagnosis classifier notebook.

```python
# ── 1. Setup and Imports ──────────────────────────────────────────────────

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import joblib

from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing

plt.rcParams['figure.figsize'] = (12, 5)
sns.set_theme(style='whitegrid')
print('All imports successful.')
```

```python
# ── 2. Load and Inspect Data ──────────────────────────────────────────────

df_raw = pd.read_csv('hhs.csv')
print('Dataset shape:', df_raw.shape)
df_raw.head(10)
```

```python
# ── 3. Data Cleaning and Preprocessing ───────────────────────────────────

df = df_raw.copy()

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%B %d, %Y')

# Remove commas from numeric columns and coerce to float
num_cols = [c for c in df.columns if c != 'Date']
for c in num_cols:
    df[c] = df[c].astype(str).str.replace(',', '', regex=False)
    df[c] = pd.to_numeric(df[c], errors='coerce')

# Rename columns for convenience
df = df.rename(columns={
    'Children apprehended and placed in CBP custody*': 'apprehended_cbp',
    'Children in CBP custody':                         'in_cbp_custody',
    'Children transferred out of CBP custody':         'transferred_out_cbp',
    'Children in HHS Care':                            'in_hhs_care',
    'Children discharged from HHS Care':               'discharged'
})

# Sort, drop duplicates, set index
df = df.sort_values('Date').drop_duplicates(subset='Date', keep='last').set_index('Date')

# Reindex to continuous daily dates; interpolate gaps
full_idx = pd.date_range(start=df.index.min(), end=df.index.max(), freq='D')
df = df.reindex(full_idx).interpolate(method='linear').ffill().bfill()
df.index.name = 'Date'

print('Final shape after cleaning:', df.shape)
df.head()
```

```python
# Drop unnecessary columns if present (mirrors dropping 'id'/'Unnamed:32' in reference notebook)
drop_candidates = ['Unnamed: 0']
df.drop(columns=[c for c in drop_candidates if c in df.columns], inplace=True)
df
```

```python
# ── 5. Feature Engineering ────────────────────────────────────────────────

data = df.copy()

# Net pressure: transfers in minus discharges
data['net_pressure'] = data['transferred_out_cbp'] - data['discharged']

# Lag features (1, 7, 14 days)
for col in ['in_hhs_care', 'discharged', 'net_pressure']:
    for lag in [1, 7, 14]:
        data[f'{col}_lag{lag}'] = data[col].shift(lag)

# Rolling mean and std (7 and 14 days, shifted by 1 to avoid leakage)
for col in ['in_hhs_care', 'discharged', 'net_pressure']:
    for window in [7, 14]:
        data[f'{col}_roll_mean{window}'] = data[col].shift(1).rolling(window).mean()
        data[f'{col}_roll_std{window}']  = data[col].shift(1).rolling(window).std()

# Calendar features
data['day_of_week'] = data.index.dayofweek
data['month']       = data.index.month
data['is_weekend']  = (data['day_of_week'] >= 5).astype(int)

data = data.dropna()
print('Feature set shape:', data.shape)
data.head()
```

```python
# ── 6. Train-Test Split (Time-Based) ─────────────────────────────────────
# Mirror: train_test_split(X, Y, test_size=0.2, random_state=42)
# For time-series we use the last 14 days as the held-out test set.

horizon = 14
train = data.iloc[:-horizon]
test  = data.iloc[-horizon:]

# Define feature columns (exclude raw targets to prevent leakage)
TARGET_COLS = ['in_hhs_care', 'discharged', 'apprehended_cbp',
               'in_cbp_custody', 'transferred_out_cbp']
feature_cols = [c for c in data.columns if c not in TARGET_COLS]

X_train = train[feature_cols]
X_test  = test[feature_cols]
y_train_care  = train['in_hhs_care']
y_test_care   = test['in_hhs_care']
y_train_disch = train['discharged']
y_test_disch  = test['discharged']

print(f'Training period : {train.index.min().date()} → {train.index.max().date()}')
print(f'Test period     : {test.index.min().date()}  → {test.index.max().date()}')
print(f'Training set shape : {X_train.shape}')
print(f'Testing  set shape : {X_test.shape}')

# Walk-forward CV splits (used in hyperparameter tuning)
tscv = TimeSeriesSplit(n_splits=5)
```

```python
# ── 7. Preprocessing Pipeline ────────────────────────────────────────────
# Mirrors ColumnTransformer + StandardScaler from reference notebook.
# All features are numeric (no categorical columns after engineering),
# so we use a simple StandardScaler within each ML pipeline below.

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_feature_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_feature_cols = X_train.select_dtypes(include=['object']).columns.tolist()

print(f'Numeric features : {len(num_feature_cols)}')
print(f'Categorical features: {len(cat_feature_cols)}')

num_transformer = StandardScaler()
cat_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer([
    ('Num', num_transformer, num_feature_cols),
    ('Cat', cat_transformer, cat_feature_cols)
], remainder='passthrough')

print('Preprocessor defined.')
```

```python
# ── 8. Baseline Models ────────────────────────────────────────────────────

# Naive persistence: repeat last observed value
naive_forecast = np.repeat(y_train_care.iloc[-1], horizon)

# Moving average: mean of last 7 days
ma_forecast = np.repeat(y_train_care.tail(7).mean(), horizon)

print('Naive forecast  (first 5):', naive_forecast[:5])
print('Moving-average forecast   :', ma_forecast[0])
```

```python
# ── 9. Statistical Models ─────────────────────────────────────────────────

# ARIMA(2,1,2)
arima_model = ARIMA(y_train_care, order=(2, 1, 2))
arima_fit   = arima_model.fit()
arima_forecast = arima_fit.forecast(steps=horizon)
print(arima_fit.summary())
```

```python
# Holt-Winters Exponential Smoothing (additive trend + weekly seasonality)
hw_model   = ExponentialSmoothing(y_train_care, trend='add',
                                   seasonal='add', seasonal_periods=7)
hw_fit      = hw_model.fit()
hw_forecast = hw_fit.forecast(steps=horizon)
print('Exponential Smoothing forecast (first 5):', hw_forecast.values[:5])
```

```python
# ── 10. Machine Learning Models (Pipelines) ───────────────────────────────
# Mirrors: Pipeline([('Pre', preprocessor), ('LogR', KNeighborsClassifier(...))])
# Here we use RF and GBR regressors within the same Pipeline pattern.

# Random Forest Pipeline
rf_pipeline = Pipeline([
    ('Pre',   preprocessor),
    ('model', RandomForestRegressor(random_state=42))
])

# Gradient Boosting Pipeline
gbr_pipeline = Pipeline([
    ('Pre',   preprocessor),
    ('model', GradientBoostingRegressor(random_state=42))
])

# Fit and predict
rf_pipeline.fit(X_train, y_train_care)
rf_forecast = rf_pipeline.predict(X_test)

gbr_pipeline.fit(X_train, y_train_care)
gbr_forecast = gbr_pipeline.predict(X_test)

print('RF  forecast (first 5):', rf_forecast[:5])
print('GBR forecast (first 5):', gbr_forecast[:5])
```

```python
# ── 11. Model Evaluation ──────────────────────────────────────────────────
# Mirrors: accuracy_score + classification_report  →  MAE / RMSE / MAPE

def mape(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

def evaluate(y_true, y_pred, name):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mp   = mape(y_true, y_pred)
    print(f'[{name}]  MAE={mae:.2f}  RMSE={rmse:.2f}  MAPE={mp:.2f}%')
    return {'Model': name, 'MAE': round(mae, 2),
            'RMSE': round(rmse, 2), 'MAPE (%)': round(mp, 2)}

results = []
results.append(evaluate(y_test_care, naive_forecast,          'Naive Persistence'))
results.append(evaluate(y_test_care, ma_forecast,             'Moving Average (7d)'))
results.append(evaluate(y_test_care, arima_forecast.values,   'ARIMA(2,1,2)'))
results.append(evaluate(y_test_care, hw_forecast.values,      'Exponential Smoothing'))
results.append(evaluate(y_test_care, rf_forecast,             'Random Forest'))
results.append(evaluate(y_test_care, gbr_forecast,            'Gradient Boosting'))

results_df = pd.DataFrame(results).sort_values('RMSE').reset_index(drop=True)
results_df
```

```python
# ── 12. Forecast Visualisation ────────────────────────────────────────────
# Mirrors: plt.plot(range(1,25), scores, marker='o') accuracy curve

plt.figure(figsize=(14, 6))
plt.plot(test.index, y_test_care.values,     label='Actual',            color='black', linewidth=2)
plt.plot(test.index, naive_forecast,          '--', label='Naive')
plt.plot(test.index, ma_forecast,             '--', label='Moving Avg')
plt.plot(test.index, arima_forecast.values,   '--', label='ARIMA')
plt.plot(test.index, hw_forecast.values,      '--', label='Exp Smoothing')
plt.plot(test.index, rf_forecast,             '--', label='Random Forest')
plt.plot(test.index, gbr_forecast,            '--', label='Gradient Boosting')
plt.title('Children in HHS Care: Forecast Comparison')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ── 13. Hyperparameter Tuning (Walk-Forward CV) ───────────────────────────
# Mirrors: for k in range(1,25): KNeighborsClassifier(n_neighbors=k) loop

# ─ Random Forest ─
rf_param_grid = {
    'model__n_estimators':    [100, 200],
    'model__max_depth':       [5, 10, None],
    'model__min_samples_leaf':[1, 2, 4]
}
rf_grid = GridSearchCV(rf_pipeline, rf_param_grid, cv=tscv,
                       scoring='neg_root_mean_squared_error', n_jobs=-1)
rf_grid.fit(X_train, y_train_care)
print('Best RF params :', rf_grid.best_params_)
print('Best RF CV RMSE:', -rf_grid.best_score_)

# ─ Gradient Boosting ─
gbr_param_grid = {
    'model__n_estimators':  [100, 200],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__max_depth':     [2, 3, 4]
}
gbr_grid = GridSearchCV(gbr_pipeline, gbr_param_grid, cv=tscv,
                        scoring='neg_root_mean_squared_error', n_jobs=-1)
gbr_grid.fit(X_train, y_train_care)
print('Best GBR params :', gbr_grid.best_params_)
print('Best GBR CV RMSE:', -gbr_grid.best_score_)
```

```python
# Evaluate tuned models on test set
best_rf_forecast  = rf_grid.best_estimator_.predict(X_test)
best_gbr_forecast = gbr_grid.best_estimator_.predict(X_test)

tuned_results = []
tuned_results.append(evaluate(y_test_care, best_rf_forecast,  'Tuned RF'))
tuned_results.append(evaluate(y_test_care, best_gbr_forecast, 'Tuned GBR'))

final_results = pd.concat(
    [results_df, pd.DataFrame(tuned_results)], ignore_index=True
).sort_values('RMSE').reset_index(drop=True)
final_results
```

```python
# ── 14. Cross-Validation (Walk-Forward) ───────────────────────────────────
# Mirrors: cross_val_score(m, X, y, cv=5, scoring='f1_weighted')

best_model_name = final_results.iloc[0]['Model']
best_estimator  = rf_grid.best_estimator_ if 'RF' in best_model_name else gbr_grid.best_estimator_

cv_scores = cross_val_score(
    best_estimator,
    X_train,
    y_train_care,
    cv=tscv,
    scoring='neg_root_mean_squared_error'
)

cv_rmse = -cv_scores
print(f'Walk-Forward CV RMSE scores : {cv_rmse.round(2)}')
print(f'Mean CV RMSE                : {cv_rmse.mean():.2f}')
print(f'Std  CV RMSE                : {cv_rmse.std():.2f}')

# Plot CV scores across folds (mirrors k-vs-accuracy plot)
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(cv_rmse)+1), cv_rmse, marker='o', color='steelblue')
plt.xlabel('Fold')
plt.ylabel('RMSE')
plt.title(f'Walk-Forward CV RMSE — {best_model_name}')
plt.tight_layout()
plt.show()
```

```python
# ── 15. Forecast Discharge Demand (Placement Needs) ──────────────────────

best_gbr_params = {k.replace('model__', ''): v
                   for k, v in gbr_grid.best_params_.items()}

gbr_disch = Pipeline([
    ('Pre',   preprocessor),
    ('model', GradientBoostingRegressor(random_state=42, **best_gbr_params))
])
gbr_disch.fit(X_train, y_train_disch)
discharge_forecast = gbr_disch.predict(X_test)

disch_eval = evaluate(y_test_disch, discharge_forecast,
                      'Gradient Boosting (Discharges)')
print(disch_eval)

plt.figure(figsize=(14, 5))
plt.plot(test.index, y_test_disch.values, label='Actual Discharges',
         color='black', linewidth=2)
plt.plot(test.index, discharge_forecast, '--',
         label='Forecasted Discharges', color='darkorange')
plt.title('Discharge / Placement Demand: Forecast vs Actual')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ── 16. Key Performance Indicators (KPIs) ─────────────────────────────────

best_model_row = final_results.iloc[0]

# ① Forecast Accuracy
forecast_accuracy = 100 - best_model_row['MAPE (%)']

# ② Surge Lead Time: days with positive net pressure in the test horizon
net_pressure_test = data['net_pressure'].iloc[-horizon:].values
surge_days = (net_pressure_test > 0).sum()

# ③ Capacity Breach Probability (95th-percentile historical load as threshold)
capacity_threshold = data['in_hhs_care'].quantile(0.95)
chosen_forecast    = best_rf_forecast if 'RF' in best_model_row['Model'] else best_gbr_forecast
breach_prob        = np.mean(chosen_forecast > capacity_threshold)

# ④ Forecast Stability Index
stability_index = 1 - (np.std(chosen_forecast - y_test_care.values)
                        / np.std(y_test_care.values))

print('=' * 45)
print('     KEY PERFORMANCE INDICATORS')
print('=' * 45)
print(f'Best Model                   : {best_model_row["Model"]}')
print(f'Forecast Accuracy            : {forecast_accuracy:.2f}%')
print(f'Surge Lead-Time Days         : {surge_days} / {horizon}')
print(f'Capacity Breach Probability  : {breach_prob:.2%}  (threshold={capacity_threshold:.0f})')
print(f'Forecast Stability Index     : {stability_index:.3f}')
print('=' * 45)
```

```python
# ── 17. Save Best Model ───────────────────────────────────────────────────
# Mirrors: joblib.dump(pipe, 'model.pkl')

best_pipeline = rf_grid.best_estimator_ if 'RF' in best_model_row['Model'] else gbr_grid.best_estimator_

joblib.dump(best_pipeline, 'best_uac_model.pkl')
joblib.dump(gbr_disch,     'discharge_model.pkl')

print('Best care-load model saved  → best_uac_model.pkl')
print('Discharge demand model saved→ discharge_model.pkl')
```
