```python
# ## 1. Setup and Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import joblib

from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing

plt.rcParams['figure.figsize'] = (12, 5)
```

```python
# ## 2. Load and Inspect Data

df_raw = pd.read_csv('hhs.csv')
print("Dataset shape:", df_raw.shape)
df_raw.head()
```

```python
df_raw.info()
```

```python
# ## 3. Data Cleaning and Preprocessing

# - Convert Date to datetime
df = df_raw.copy()
df['Date'] = pd.to_datetime(df['Date'], format='%B %d, %Y')

# - Remove commas from numeric columns
num_cols = [c for c in df.columns if c != 'Date']
for c in num_cols:
    df[c] = df[c].astype(str).str.replace(',', '', regex=False)
    df[c] = pd.to_numeric(df[c], errors='coerce')

# - Rename columns for convenience
df = df.rename(columns={
    'Children apprehended and placed in CBP custody*': 'apprehended_cbp',
    'Children in CBP custody': 'in_cbp_custody',
    'Children transferred out of CBP custody': 'transferred_out_cbp',
    'Children in HHS Care': 'in_hhs_care',
    'Children discharged from HHS Care': 'discharged'
})

# - Sort by date and set index
df = df.sort_values('Date').drop_duplicates(subset='Date', keep='last').set_index('Date')

# Reindex to continuous daily dates interpolate missing days
full_idx = pd.date_range(start=df.index.min(), end=df.index.max(), freq='D')
df = df.reindex(full_idx)
df = df.interpolate(method='linear').ffill().bfill()
df.index.name = 'Date'

print("Final shape after cleaning:", df.shape)
df.head()
```

```python
# ## 4. Exploratory Data Analysis

fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)
df['in_hhs_care'].plot(ax=axes[0], title='Children in HHS Care over time', color='steelblue')
df['discharged'].plot(ax=axes[1], title='Children Discharged from HHS Care (daily)', color='darkorange')
plt.tight_layout()
plt.show()
```

```python
# Seasonal decomposition of care load (weekly seasonality)
decomposition = seasonal_decompose(df['in_hhs_care'], model='additive', period=7)
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.tight_layout()
plt.show()
```

```python
# ## 5. Feature Engineering

# - Net pressure indicator: Transfers - Discharges
data = df.copy()
data['net_pressure'] = data['transferred_out_cbp'] - data['discharged']

# - Lag features (1, 7, 14 days)
for col in ['in_hhs_care', 'discharged', 'net_pressure']:
    for lag in [1, 7, 14]:
        data[f'{col}_lag{lag}'] = data[col].shift(lag)

# - Rolling statistics (mean and std over 7 and 14 days)
for col in ['in_hhs_care', 'discharged', 'net_pressure']:
    for window in [7, 14]:
        data[f'{col}_roll_mean{window}'] = data[col].shift(1).rolling(window).mean()
        data[f'{col}_roll_std{window}'] = data[col].shift(1).rolling(window).std()

# - Calendar features (day of week, month, weekend)
data['day_of_week'] = data.index.dayofweek
data['month'] = data.index.month
data['is_weekend'] = (data['day_of_week'] >= 5).astype(int)

data = data.dropna()
print("Feature set shape:", data.shape)
data.head()
```

```python
# ## 6. Train-Test Split (Time‑Based)

# We use the last `horizon = 14` days as the test set.
horizon = 14
train = data.iloc[:-horizon]
test = data.iloc[-horizon:]

feature_cols = [c for c in data.columns if c not in 
                ['in_hhs_care', 'discharged', 'apprehended_cbp', 'in_cbp_custody', 'transferred_out_cbp']]

X_train, y_train_care = train[feature_cols], train['in_hhs_care']
X_test, y_test_care = test[feature_cols], test['in_hhs_care']
y_train_disch, y_test_disch = train['discharged'], test['discharged']

print(f"Training period: {train.index.min().date()} to {train.index.max().date()}")
print(f"Test period: {test.index.min().date()} to {test.index.max().date()}")

# Walk-forward cross-validation splits
tscv = TimeSeriesSplit(n_splits=5)
```

```python
# ## 7. Baseline Models

# - Naive persistence (last observed value)
naive_forecast = np.repeat(y_train_care.iloc[-1], horizon)

# - Moving average (last 7 days)
ma_value = y_train_care.tail(7).mean()
ma_forecast = np.repeat(ma_value, horizon)

print("Naive forecast (first 5):", naive_forecast[:5])
print("Moving average forecast:", ma_value)
```

```python
# ## 8. Statistical Models

# ARIMA(2,1,2) – chosen based on ACF/PACF exploration (simplified)
arima_model = ARIMA(y_train_care, order=(2, 1, 2))
arima_fit = arima_model.fit()
arima_forecast = arima_fit.forecast(steps=horizon)
print(arima_fit.summary())
```

```python
# Exponential Smoothing with additive trend and weekly seasonality
hw_model = ExponentialSmoothing(y_train_care, trend='add', seasonal='add', seasonal_periods=7)
hw_fit = hw_model.fit()
hw_forecast = hw_fit.forecast(steps=horizon)
print("Exponential Smoothing forecast (first 5):", hw_forecast[:5])
```

```python
# ## 9. Machine Learning Models (Pipelines)

# Random Forest Pipeline
rf_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestRegressor(random_state=42))
])

# Gradient Boosting Pipeline
gbr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', GradientBoostingRegressor(random_state=42))
])

# Train and predict
rf_pipeline.fit(X_train, y_train_care)
rf_forecast = rf_pipeline.predict(X_test)

gbr_pipeline.fit(X_train, y_train_care)
gbr_forecast = gbr_pipeline.predict(X_test)

print("RF forecast (first 5):", rf_forecast[:5])
print("GBR forecast (first 5):", gbr_forecast[:5])
```

```python
# ## 10. Model Evaluation

def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluate(y_true, y_pred, name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mp = mape(y_true, y_pred)
    return {'Model': name, 'MAE': mae, 'RMSE': rmse, 'MAPE (%)': mp}

results = []
results.append(evaluate(y_test_care, naive_forecast, 'Naive Persistence'))
results.append(evaluate(y_test_care, ma_forecast, 'Moving Average (7d)'))
results.append(evaluate(y_test_care, arima_forecast.values, 'ARIMA(2,1,2)'))
results.append(evaluate(y_test_care, hw_forecast.values, 'Exponential Smoothing'))
results.append(evaluate(y_test_care, rf_forecast, 'Random Forest'))
results.append(evaluate(y_test_care, gbr_forecast, 'Gradient Boosting'))

results_df = pd.DataFrame(results).sort_values('RMSE')
results_df
```

```python
# Visualization of forecasts vs actual
plt.figure(figsize=(14, 6))
plt.plot(test.index, y_test_care.values, label='Actual', color='black', linewidth=2)
plt.plot(test.index, naive_forecast, '--', label='Naive')
plt.plot(test.index, ma_forecast, '--', label='Moving Avg')
plt.plot(test.index, arima_forecast.values, '--', label='ARIMA')
plt.plot(test.index, hw_forecast.values, '--', label='Exp Smoothing')
plt.plot(test.index, rf_forecast, '--', label='Random Forest')
plt.plot(test.index, gbr_forecast, '--', label='Gradient Boosting')
plt.title('Children in HHS Care: Forecast Comparison')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ## 11. Hyperparameter Tuning (Walk‑Forward CV)

# Tuning Random Forest
rf_param_grid = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [5, 10, None],
    'model__min_samples_leaf': [1, 2, 4]
}
rf_grid = GridSearchCV(rf_pipeline, rf_param_grid, cv=tscv,
                        scoring='neg_root_mean_squared_error', n_jobs=-1)
rf_grid.fit(X_train, y_train_care)
print("Best RF params:", rf_grid.best_params_)
print("Best RF CV RMSE:", -rf_grid.best_score_)

# Tuning Gradient Boosting
gbr_param_grid = {
    'model__n_estimators': [100, 200],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__max_depth': [2, 3, 4]
}
gbr_grid = GridSearchCV(gbr_pipeline, gbr_param_grid, cv=tscv,
                         scoring='neg_root_mean_squared_error', n_jobs=-1)
gbr_grid.fit(X_train, y_train_care)
print("Best GBR params:", gbr_grid.best_params_)
print("Best GBR CV RMSE:", -gbr_grid.best_score_)

# Evaluate tuned models on test set
best_rf_forecast = rf_grid.best_estimator_.predict(X_test)
best_gbr_forecast = gbr_grid.best_estimator_.predict(X_test)

tuned_results = []
tuned_results.append(evaluate(y_test_care, best_rf_forecast, 'Tuned RF'))
tuned_results.append(evaluate(y_test_care, best_gbr_forecast, 'Tuned GBR'))
final_results = pd.concat([results_df, pd.DataFrame(tuned_results)], ignore_index=True).sort_values('RMSE')
final_results
```

```python
# ## 12. Forecast Discharge Demand (Placement Needs)

# Use best GBR hyperparameters for discharge forecasting
gbr_disch = Pipeline([
    ('scaler', StandardScaler()),
    ('model', GradientBoostingRegressor(random_state=42, 
                                        **{k.replace('model__',''): v for k, v in gbr_grid.best_params_.items()}))
])
gbr_disch.fit(X_train, y_train_disch)
discharge_forecast = gbr_disch.predict(X_test)

disch_eval = evaluate(y_test_disch, discharge_forecast, 'Gradient Boosting (Discharges)')
print(disch_eval)

plt.figure(figsize=(14, 5))
plt.plot(test.index, y_test_disch.values, label='Actual Discharges', color='black', linewidth=2)
plt.plot(test.index, discharge_forecast, '--', label='Forecasted Discharges', color='darkorange')
plt.title('Discharge / Placement Demand: Forecast vs Actual')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ## 13. Key Performance Indicators (KPIs)
# - Forecast Accuracy (%) = 100 - MAPE
# - Surge Lead Time: days with positive net pressure
# - Capacity Breach Probability (% of days forecast > threshold)
# - Forecast Stability Index: 1 - (RMSE / std(actual))

best_model_row = final_results.iloc[0]
forecast_accuracy = 100 - best_model_row['MAPE (%)']

net_pressure_forecast = data['net_pressure'].tail(horizon).values
surge_days = (net_pressure_forecast > 0).sum()

capacity_threshold = data['in_hhs_care'].quantile(0.95)  # 95th percentile historical load
breach_prob = np.mean(rf_forecast > capacity_threshold) if 'Random Forest' in best_model_row['Model'] else np.mean(gbr_forecast > capacity_threshold)

stability_index = 1 - (np.std(best_rf_forecast - y_test_care.values) / np.std(y_test_care.values))

print("===== Key Performance Indicators =====")
print(f"Best Model: {best_model_row['Model']}")
print(f"Forecast Accuracy: {forecast_accuracy:.2f}%")
print(f"Surge Lead Time (positive net pressure days in horizon): {surge_days} / {horizon}")
print(f"Capacity Breach Probability (threshold={capacity_threshold:.0f}): {breach_prob:.2%}")
print(f"Forecast Stability Index: {stability_index:.3f}")
```
